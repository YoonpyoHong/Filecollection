```python
import matplotlib.pyplot as plt
import csv
f = open('seoul10.csv')
data = csv.reader(f)
next(data)
result_high = []
result_low = []
for row in data:
    if row[-1] != '' and row[-2] != '':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '31':
            result_high.append(float(row[-1]))
            result_low.append(float(row[-2]))
plt.plot(result_high, label = 'high')
plt.plot(result_low,label = 'low')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul10.csv')
data = csv.reader(f)
next(data)
result_jan = []
result_aug = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '01':
            result_jan.append(float(row[-1]))
        elif row[0].split('-')[1] == '08':
            result_aug.append(float(row[-1]))
plt.hist(result_jan,bins = 100, label = 'Jan')
plt.hist(result_aug,bins = 100, label = 'Aug')
plt.legend()
plt.show()
```
