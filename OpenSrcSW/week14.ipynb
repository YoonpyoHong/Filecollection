```python
import sys
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np

f = open('people.csv')
data = csv.reader(f)
next(data)

people = [[] for _ in range(2)]

for row in data:
    if row[0] == '서울특별시 광진구 (1121500000)':
        people[0] = row[3:14]
        
f.close()
f = open('people.csv')
data = csv.reader(f)
next(data)

result = 0
result_data = []

for row in data:
    temp = 100000000000
    i = 0
    for data_read in row[3:14]:
     
        temp += int(people[0][i].replace(",", ""))^2 - int(data_read.replace(",", ""))^2
        i += 1
    if temp > result:
        result_data = row[3:14]
people[1]= result_data  
for i in range(len(people[0])):
    people[0][i] = int(people[0][i].replace(",", ""))
    people[1][i] = int(people[1][i].replace(",", ""))
      


plt.style.use('ggplot')
plt.plot(people[0])
plt.plot(people[1])
plt.show()
```
